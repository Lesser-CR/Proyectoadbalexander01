# Revoke permissions to schemas and catalog

```python
%sql
REVOKE ALL PRIVILEGES ON CATALOG chocolate_sales FROM `Data_Engineers`;
REVOKE CREATE, USE SCHEMA ON SCHEMA  chocolate_sales.raw FROM `Data_Engineers`;
REVOKE CREATE, USE SCHEMA ON SCHEMA  chocolate_sales.bronze FROM `Data_Engineers`;
REVOKE CREATE, USE SCHEMA ON SCHEMA  chocolate_sales.silver FROM `Data_Engineers`;
REVOKE CREATE, USE SCHEMA ON SCHEMA  chocolate_sales.golden FROM `Data_Engineers`;
```

# Revoke permissions to external locations

```python
%sql
REVOKE READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-bronze` FROM `Data_Engineers`;
REVOKE READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-silver` FROM `Data_Engineers`;
REVOKE READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-golden` FROM `Data_Engineers`;
REVOKE READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-raw` FROM `Data_Engineers`;
```
