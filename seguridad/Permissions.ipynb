# Grant permissions to schemas and catalog

```python
%sql
GRANT ALL PRIVILEGES ON CATALOG chocolate_sales TO `Data_Engineers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  chocolate_sales.raw to `Data_Engineers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  chocolate_sales.bronze to `Data_Engineers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  chocolate_sales.silver to `Data_Engineers`;
GRANT CREATE, USE SCHEMA ON SCHEMA  chocolate_sales.golden to `Data_Engineers`;
```

# Grant permissions to external locations

```python
%sql
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-bronze` to `Data_Engineers`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-silver` to `Data_Engineers`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-golden` to `Data_Engineers`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-raw` to `Data_Engineers`;
```
